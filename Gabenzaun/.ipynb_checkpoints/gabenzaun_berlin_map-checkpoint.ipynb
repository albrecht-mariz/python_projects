```python
import pandas as pd
import folium
```

```python
#Berlin coordinates for folium
Berlin = [52.520008, 13.404954]
```

```python
df = pd.read_excel('data/liste_raw.xlsx')
```

```python
df[['lat','lon']] = df.google_coord.str.split(",",expand=True)
```

```python
df['count'] = 1 
```

```python
df.head(5)
```

```python
map_ = folium.Map(location=Berlin, zoom_start=12, tiles = "cartodbpositron", control_scale = True) #zoom_control = False

df.apply(lambda row:folium.CircleMarker(location=[row["lat"], row["lon"]], 
                                              radius=6, stroke=False, fill_color='green', fill_opacity=0.7)
                                             .add_to(map_), axis=1)

map_
```

```python
map_.save('map/gabenzaun_map.html')
```
